```python
from transformers import pipeline
```

```python
pipe = pipeline(task="text-classification")
```

```python
pipe(
    [
        "I love the new design of your website, it's so user-friendly!",
        "This is the worst customer service experience I've ever had.",
        "The movie was absolutely fantastic, I enjoyed every moment of it.",
        "I am extremely disappointed with the quality of the product.",
        "What a beautiful day! I'm so happy to be outside.",
        "The presentation was incredibly boring and put me to sleep.",
        "I feel really positive about the changes in our company.",
        "The food at the restaurant was terrible, I will not be going back.",
        "I'm grateful for all the support and encouragement from my friends.",
        "The traffic this morning was horrendous, made me late for work.",
    ]
)
```

```python
pipe = pipeline(
    "token-classification",
    model="Clinical-AI-Apollo/Medical-NER",
    aggregation_strategy="simple",
)
result = pipe("63 year old woman with history of CAD presented to ER")
```
